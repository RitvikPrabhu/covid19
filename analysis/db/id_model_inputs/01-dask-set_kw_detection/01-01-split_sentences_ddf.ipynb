```python
import gc
import pathlib as pl
import sys

import dask.dataframe as dd
from dask.distributed import Client
import pandas as pd
from pyprojroot import here
```

```python
# creating local cluster, LocalCluster
client = Client(n_workers=3, threads_per_worker=1, memory_limit='2GB')
```

```python
client
```

```python
paper_df = pd.read_csv(here("./data/db/final/kaggle/paper_text/document_parses_pmc_json.tsv"),
                       sep = "\t")
```

```python
nparts = 1000
ddf = dd.from_pandas(paper_df, npartitions=nparts)
```

```python
print(sys.getrefcount(paper_df))
del paper_df
```

```python
gc.collect()
```

```python
pl.Path(here("./data/db/working/kaggle/id_model_inputs", warn=False)).mkdir(parents=True, exist_ok=True)
fout = str(here(f"./data/db/working/kaggle/id_model_inputs/01-01-dask_data_frame_partitions_{nparts}.json.gzip", warn=False))
fout
```

```python
ddf.to_json(fout, compression="gzip") # 2 minutes
```

```python
client.shutdown()
```
